```python
webdriver_path = 'C:/Users/yangh/Documents/chromedriver_win32/chromedriver'
driver = webdriver.Chrome(webdriver_path)

url='https://klue.kr/'
driver.get(url)
```

```python
temp_num=len(driver.find_elements_by_class_name("lecture-search-result"))
temp_num
```

```python
temp_num_1=len(driver.find_elements_by_class_name("lecture-search-result"))
temp_num_1

for i in range(temp_num_1):
    temp_xpath="""/html/body/app-root/lecture-result/div/lecture-list/div/div["""+str(i+3)+"""]/div[2]/div[2]/a/button"""
    driver.find_element_by_xpath(temp_xpath).click()
    temp_num_2=len(driver.find_elements_by_class_name("lecture-detail-content"))
    for j in range(temp_num_2):
        temp_xpath_2="""/html/body/app-root/lecture-view/div/lecture-eval/div/div["""+str(j+2)+"""]/div[1]/div[2]"""
        print(driver.find_element_by_xpath(temp_xpath_2).text)
    print('\n')
    driver.execute_script("window.history.go(-1)")
```

```python
col=['코드명+교수명','과목명','교수명','과목코드','이수구분','개설학기','학점','ID','총점','출석체크','학점','난이도','학습량','성취감','강의평']
data_list=[]
#len(df.index)
for i in range(len(df.index)):
    course=df.iloc[i,6]+' '+df.iloc[i,11]
    temp_url="https://klue.kr/lecture/search/"+df.iloc[i,6]+"%20"+df.iloc[i,11]
    driver.implicitly_wait(2)
    driver.get(temp_url)
    
    temp_num_1=len(driver.find_elements_by_class_name("lecture-search-result"))
    temp_num_1

    for i in range(temp_num_1):
        temp_xpath="""/html/body/app-root/lecture-result/div/lecture-list/div/div["""+str(i+3)+"""]/div[2]/div[2]/a/button"""
        driver.find_element_by_xpath(temp_xpath).click()
        new_input=[]
        new_input.append(course)
        #과목명
        new_input.append(driver.find_element_by_xpath('/html/body/app-root/lecture-view/div/lecture-info/div/div[1]/div[2]/div[1]').text)
        #교수명
        new_input.append(driver.find_element_by_xpath('/html/body/app-root/lecture-view/div/lecture-info/div/div[1]/div[2]/div[2]').text)
        #과목코드
        new_input.append(driver.find_element_by_xpath('/html/body/app-root/lecture-view/div/lecture-info/div/div[1]/div[2]/div[4]/table/tbody/tr[1]/td[2]').text)
        #이수구분
        new_input.append(driver.find_element_by_xpath('/html/body/app-root/lecture-view/div/lecture-info/div/div[1]/div[2]/div[4]/table/tbody/tr[2]/td[1]').text)
        #개설학기
        new_input.append(driver.find_element_by_xpath('/html/body/app-root/lecture-view/div/lecture-info/div/div[1]/div[2]/div[4]/table/tbody/tr[1]/td[1]').text)
        #학점
        new_input.append(driver.find_element_by_xpath('/html/body/app-root/lecture-view/div/lecture-info/div/div[1]/div[2]/div[4]/table/tbody/tr[3]/td[1]').text)
        #ID
        new_input.append("평균")
        #총점
        new_input.append(driver.find_element_by_xpath("""/html/body/app-root/lecture-view/div/lecture-info/div/div[1]/div[2]/div[3]/div/span[2]""").text)

        a=driver.find_element_by_tag_name('g')
        #출석체크
        new_input.append(a.find_elements_by_tag_name('text')[5].text)
        #학점
        new_input.append(a.find_elements_by_tag_name('text')[6].text)
        #난이도
        new_input.append(a.find_elements_by_tag_name('text')[7].text)
        #학습량
        new_input.append(a.find_elements_by_tag_name('text')[8].text)
        #성취감
        new_input.append(a.find_elements_by_tag_name('text')[9].text)
        #강의평
        new_input.append('-')

        data_list.append(new_input)
        driver.execute_script("window.history.go(-1)")

new_df=pd.DataFrame(data_list, columns=col)
```
